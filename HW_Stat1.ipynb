Исходя из данных большинство животных имеют повышенную температуру, наименьшее число животных имеют пониженную или нормальную температуру,что говорит о возможной наличии инфекции у большинства животных. Большинство лошадей имеют повышенный пульс, что говорит о возможном состоянии шока или болезни животного, при этом случаев с пониженным пульсом нет(относительно низкие показатели могут встречаться у спортивных животных). Частота дыхания у большей части животных выше нормы

Большинство лошадей по субъективным оценкам испытывают легкую прерывистую боль, что может говорить о том, что на момент осмотра были приняты меры по лечению, или(в более редких случаях, т.к. большинство животных прооперировано) пониженную вероятность в необходимости в операции.
```python
q1 = df['rectal_temperature'].quantile(0.25)
q3 = df['rectal_temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['rectal_temperature'].between(lower_bound, upper_bound, inclusive=True)]
remove_outliers
```

```python
pd.concat([df, remove_outliers]).drop_duplicates(keep=False) #При инфекции температура лошади может поднматься до 42 гр., очень низкая температура может быть в состоянии шока
```

```python
q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['pulse'].between(lower_bound, upper_bound, inclusive=True)]
remove_outliers
```

```python
pd.concat([df, remove_outliers]).drop_duplicates(keep=False) #140 -- не наблюдается даже у жеребят в первые недели жизни, но в сочетании с остальными критическими параметрами и последующей эвтаназией можно предположить, что это не выброс. Остальные данные -- пропуски
```
